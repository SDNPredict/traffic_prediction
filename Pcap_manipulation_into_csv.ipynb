```python
# This script reads Wireshark-converted CSV files containing network traffic data (pcap files).
# It extracts the first two columns (typically packet number and timestamp)
# and saves them to a new file.

import csv

array_files = ['./dumps/h1_mixed_31_03.csv', './dumps/h1_mixed_31_03.csv', './dumps/h1_mixed_31_03.csv', './dumps/h1_mixed_31_03.csv']

for index, (file) in enumerate(array_files, 1):
    out_file_name = './dumps/h' + str(index) + '_mixed_31_03_filtered.csv'
    with open(file) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        line_count = 0
        with open(out_file_name, 'w+') as csv_formatted_file:
            for row in csv_reader:
                if line_count == 0:
                    print(f'Column names are {", ".join(row)}')
                    line_count += 1
                else:
                    csv_formatted_file.write(f"{row[0]},{row[1]}\n")
            
```

```python
import numpy as np

def triangular_wave(t, period):
    """
    This function generates a triangular wave with the given time (t) and period
    """
    amplitude = 1
    if t < 0:
        raise ValueError("Time (t) must be positive")

    triangular_value = abs((4 * amplitude / period) * abs(((t - (period / 4)) % period) - (period / 2)) - amplitude)

    return triangular_value


period = 1  # seconds
amplitude = 0.5
time_points = np.linspace(0, period, 1000)  # 1000 points within the period

wave_values = [triangular_wave(t, period) for t in time_points]

# Plot the wave (using matplotlib)
import matplotlib.pyplot as plt

plt.plot(time_points, wave_values)
plt.xlabel("Time (t)")
plt.ylabel("Amplitude")
plt.title("Triangular Wave")
plt.show()
```
