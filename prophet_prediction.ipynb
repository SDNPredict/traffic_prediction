```python
# Python
import pandas as pd
from prophet import Prophet
import os
```

```python
# Python
# df = pd.read_csv('https://raw.githubusercontent.com/facebook/prophet/main/examples/example_yosemite_temps.csv')
FP = "assets/demo-2-4-aprile/h4_31_03_filtered_converted_stretched.csv"
df = pd.read_csv(FP)
m = Prophet(changepoint_prior_scale=0.01).fit(df)
```

```python
future = m.make_future_dataframe(periods=120, freq='H')
fcst = m.predict(future)
fig = m.plot(fcst)
```
