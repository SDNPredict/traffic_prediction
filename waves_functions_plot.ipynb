```python
import numpy as np
import math

def triangular_wave(t, period):
    """
    This function generates a triangular wave with the given time (t) and period
    """
    amplitude = 1
    if t < 0:
        raise ValueError("Time (t) must be positive")

    triangular_value = abs((4 * amplitude / period) * abs(((t - (period / 4)) % period) - (period / 2)) - amplitude)

    return triangular_value


period = 1  # seconds
amplitude = 0.5
time_points = np.linspace(0, period, 1000)  # 1000 points within the period

wave_values = [triangular_wave(t, period) for t in time_points]

# Plot the wave (using matplotlib)
import matplotlib.pyplot as plt

plt.plot(time_points, wave_values)
plt.xlabel("Time (t)")
plt.ylabel("Amplitude")
plt.title("Triangular Wave")
plt.show()
```

```python
def sawtooth_wave(t, period):
    """
    This function generates a sawtooth wave with the given time (t) and period
    """
    amplitude = 1
    if t < 0:
        raise ValueError("Time (t) must be positive")

    sawtooth_value = 2*(t-math.floor(t+0.5)) + amplitude

    
    return sawtooth_value

period = 1  # seconds
amplitude = 0.5
time_points = np.linspace(0, 3, 1000)  # 1000 points within the period

wave_values = [sawtooth_wave(t, period) for t in time_points]

# Plot the wave (using matplotlib)
import matplotlib.pyplot as plt

plt.plot(time_points, wave_values)
plt.xlabel("Time (t)")
plt.ylabel("Amplitude")
plt.title("Sawtooth Wave")
plt.show()
```

```python
def square_wave(t, period):
    """
    This function generates a sawtooth wave with the given time (t), period, and amplitude.
    """
    amplitude = 1
    if t < 0:
        raise ValueError("Time (t) must be positive")

    square_value = 4*math.floor(t)-2*math.floor(2*t)+1

    
    return square_value

period = 1  # seconds
amplitude = 0.5
time_points = np.linspace(0, 3, 1000)  # 1000 points within the period

wave_values = [square_wave(t, period) for t in time_points]

# Plot the wave (using matplotlib)
import matplotlib.pyplot as plt

plt.plot(time_points, wave_values)
plt.xlabel("Time (t)")
plt.ylabel("Amplitude")
plt.title("Square Wave")
plt.show()
```
